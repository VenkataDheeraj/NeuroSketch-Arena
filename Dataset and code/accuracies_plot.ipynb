```python
import pandas as pd
import matplotlib.pyplot as plt

# ---------- Load Data ----------
acc_5000 = pd.read_csv('model_accuracies_5000.csv', index_col=0)
acc_10000 = pd.read_csv('model_accuracies_10000.csv', index_col=0)

# ---------- Combine into One DataFrame ----------
accuracy_df = pd.DataFrame({
    '5000 samples': acc_5000['Accuracy'],
    '10000 samples': acc_10000['Accuracy']
})

# ---------- Plot ----------
ax = accuracy_df.plot(kind='barh', figsize=(12, 8))
plt.title('Model Accuracy Comparison\n(5000 vs 10000 Samples per Class)')
plt.xlabel('Test Accuracy')
plt.ylabel('Model')
plt.grid(True, axis='x')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.tight_layout()
plt.show()
```
